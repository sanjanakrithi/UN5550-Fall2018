```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#firstly importing the csv files and printing the file. I named my file iris.
import csv
df = pd.read_csv('projects/data/iris.csv')
df.iloc[34]=4.9,3.1,1.5,0.2,"setosa"
df.iloc[37] = 4.9,3.6,1.4,0.1,"setosa"
#assign petal length to variable x1. since petal lenght is the third column we access that
x1 = df.iloc[:,2]#petal length
x1
#assign petal width to variable x2. since petal width is the fourth column we access that and print it
x2 = df.iloc[:,3]#petal width
x2
#petal raito: petal lenght/petal width
x = x1/x2#petal lenght/petal width
x
#assign sepal length to variable y1. since sepal length is the first column we access that
y1 = df.iloc[:,0]#sepal length
y1
#assign sepal width to variable y2. since sepal width is the second column we access that
y2 = df.iloc[:,1]#sepal width
y2
#"Sepal ratio" :  sepal length / sepal width"
y = y1/y2
y
df1 = df.assign(petalraito = x, sepalraito=y)
df1

pr = df1[['petalraito','Species']].groupby('Species')
sr = df1[['sepalraito','Species']].groupby('Species')
```

```python
summary = pr.describe()
summary
```

```python
#creating a boxplot using pandas 
bp = pr.boxplot(pr)
bp.set_title('Box plot showing the distribution for the three species for petal raito')
bp.set_xlabel('Petal Raito')
title_petalraito = bp.set_ylabel('Distribution of data')
```

In the above box plot shows the distribution of data for the petal raito for the three speices. There are about 50 instances for setosa, 50 instances for versicolor, 50 instances for virginica. The points which lie above or below the min and max lines are negligable since there are very few which lie out of the distribution. The sample size is much greater than the number of outliers hence they can be ignored. The green line represents the 50% of the data which colsely represents the mean. The min and max reprenst the range of the data set.We can observe that setosa is distributed from 3 to 11(range). The box plot for setosa is negatively skewed since the median is closer to the 75%. The IQR is larger compared to the other two cases which shows taht the data is distributed. Versicolor, the IQR is very small i.e., (Q3-Q1)~0.4 and the distribution is not spreadout, its rather concetrated in a given range. And the data is mostly symmentric In the case of virginica, the data is not distributed and the data is positively skewed.

A volin plot is used to show the distribution of data and probability density. The thick black lines represnt the IQR and the white dots represnt the median. The width represnts the distribution of data. If the plot is long it shows that the data is distributed over a range and the IQR is higher. Advantage of voilin plots : 1)Box plots are not effected by the distribution of data but the voiln plots show the distribution of data. 2)can be used to visualize small data sets 3)can work with both quantitative and qualitative data 4)it considers the entire set of data i.e., unlike the box plots where the outliers ae ignored, in voilin plots the entire data set is considered.
